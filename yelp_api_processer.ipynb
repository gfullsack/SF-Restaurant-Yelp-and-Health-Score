```python
# Set up containers to capture the data
#iterations = 0
restaurants = []
restaurant_count = 0

while restaurant_count < 1000:
#     iterations += 1
    try:
        print(f'Attempting to make API order... Restaurant Count: {restaurant_count}')
        payload = {
            'term': 'restaurants',
            'location': 'san francisco',
            'limit': 50,
            'offset': restaurant_count
        }

        # make the api call
        response = requests.get(uri, params=payload, headers=headers)
        print(f'Status Code: {response.status_code}')

        # grab the data
        data = response.json()
        print(f'Restaurant data: {data.keys()}')

        # add the results to our restaurants list [0].concat([1, 2, 3]) -> [0,1 2, 3, 4]
        restaurants += data['businesses']

        # keep track of the current count that we can offset the api call by
        restaurant_count += 50
    except:
        print(' Mr.Fullsack.. I dont feel so good...something went wrong in kitchen... Giving up')
        
```

```python
# check the number of dictionary objects in list
len(restaurants)
```

```python
# checking to see if the data is not duplicate
restaurant_names = []

for obj in restaurants:
    restaurant_names.append(obj['name'])
```

```python
restaurant_unique = set(restaurant_names)
print('Number of unique restaurants:', len(restaurant_unique))
```
